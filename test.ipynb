```python
tester = pd.read_excel("data/test_length.xls",
                       skiprows= 2
)

tester = tester.rename(columns= col_map)
```

```python
tester = tester.drop(columns= ['status', 'comment'])
```

```python
tester = tester.dropna()

tester
```

```python
tester['date'] = pd.to_datetime(tester['date'], format= "%d-%m-%Y")
```

```python
tester
```

```python
tester['type'] = tester['amount'].apply(lambda x : "Income" if x > 0 else "Expense")
```

```python
tester
```

```python
tester["category"].unique()
```

```python
def is_savings(row):
    if row['amount'] < 0 and row['category'] == 'Banque':
        return 'épargne'
    elif row['amount'] < 0 :
        return 'dépense'
    else:
        return 'revenu'
    
tester['label'] = tester.apply(is_savings, axis= 1)
```
